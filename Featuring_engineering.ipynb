1. What is a parameter?

   A parameter is a variable inside a machine learning model that is learned from the training data. Parameters control the behavior of the model and help it make predictions. For example, in a linear regression model, slope and intercept are parameters. These values are adjusted during training to minimize the error between predicted and actual values.

2. What is correlation?

   Correlation is a statistical measure that shows the relationship between two variables. It tells us how strongly variables are related and in which direction. The value of correlation ranges from -1 to +1, where +1 indicates strong positive correlation, -1 indicates strong negative correlation, and 0 indicates no correlation.

3. What does negative correlation mean?

   Negative correlation means that when one variable increases, the other variable decreases. For example, as the number of hours spent watching TV increases, academic performance may decrease. A negative correlation value lies between 0 and -1.

4. Define Machine Learning.

   Machine Learning is a branch of Artificial Intelligence that enables computers to learn patterns from data and make predictions or decisions without being explicitly programmed. It focuses on developing algorithms that improve automatically through experience and data.

5. What are the main components in Machine Learning?

   The main components of Machine Learning are:

   1-Data – used to train and test models

   2-Features – input variables

   3-Model – mathematical representation

   4-Loss Function – measures error

   5-Optimizer – minimizes loss

   6-Evaluation Metrics – checks performance

6. How does loss value help in determining whether the model is good or not?

   The loss value measures how far the model’s predictions are from the actual values. A lower loss value indicates better model performance, while a higher loss shows poor predictions. During training, the goal is to minimize the loss.

7. What are continuous and categorical variables?

   Continuous variables are numerical values that can take any value within a range, such as height, weight, and temperature.

Categorical variables represent categories or labels, such as gender, color, or city.

8. How do we handle categorical variables in Machine Learning? What are the     common techniques?

   Categorical variables are converted into numerical form using encoding techniques. Common techniques include:

      1-Label Encoding

      2-One-Hot Encoding

      3-Ordinal Encoding

9. What do you mean by training and testing a dataset?

      Training dataset is used to teach the model patterns from data.
      Testing dataset is used to evaluate the model’s performance on unseen data. This helps check whether the model generalizes well.

11. What is a Test set?

     A test set is a portion of the dataset used to evaluate the performance of a trained machine learning model. It helps measure how well the model performs on unseen data.

14. Why do we have to perform EDA before fitting a model to the data?

     EDA (Exploratory Data Analysis) helps understand data distribution, detect missing values, outliers, patterns, and relationships between variables. It ensures better feature selection and improves model accuracy.

16. What is causation?

     Causation means that one event directly causes another event. It indicates a cause-and-effect relationship, not just an association.

17. Explain difference between correlation and causation with an example.

     Correlation means variables move together.

     Causation means one variable causes the other.

      Example:
      Ice cream sales and drowning cases increase together (correlation), but ice cream does not cause drowning. Hot weather causes both.

18. What is an Optimizer?

      An optimizer is an algorithm used to minimize the loss function by updating model parameters during training.

19. What are different types of optimizers? Explain each with an example.

Types of optimizers:

      Gradient Descent – updates parameters using full dataset

      Stochastic Gradient Descent (SGD) – updates using one data point

      Mini-batch Gradient Descent – uses small batches

      Adam Optimizer – adaptive learning rate, fast convergence

23. What is feature scaling?

     Feature scaling is the process of transforming numerical features to a common scale so that no feature dominates the model.

24. How does feature scaling help in Machine Learning?

      Feature scaling:

      -Improves convergence speed

      -Improves accuracy

      -Helps distance-based algorithms

      -Prevents bias due to large values

25. How do we perform scaling in Python?

      Using StandardScaler:

      from sklearn.preprocessing import StandardScaler
      scaler = StandardScaler()
      X_scaled = scaler.fit_transform(X)

26. Explain data encoding.

       Data encoding converts categorical data into numerical format so that machine learning algorithms can process it. Common methods include Label Encoding and One-Hot Encoding.
